### 3.9.3 GroupBy：分割、应用和组合

#### 1.分割、应用和组合

```python
df = pd.DataFrame({'key': ['A', 'B', 'C', 'A', 'B', 'C'],
                   'data': range(6)}, columns=['key', 'data'])
df
```

可以用 DataFrame 的 groupby() 方法进行绝大多数常见的分割-应用-组合操作  
只需将需要分组的列名传进行即可

```python
df.groupby('key')
```

```python
df.groupby('key').sum()
```

#### 2.GroupBy 对象

##### (1) 按列取值

```python
planets.groupby("method")
```

```python
planets.groupby('method')['orbital_period']
```

```python
planets.groupby('method')['orbital_period'].median()
```

##### (2) 按组迭代

```python
for (method, group) in planets.groupby('method'):
    print("{0:30s} shape={1}".format(method, group.shape))
```

##### (3) 调用方法

```python
planets.head()
```

```python
planets['method'].unique()
```

```python
planets.groupby('method')['year'].describe()
```

#### 3.累计、过滤、转换和应用

```python
rng = np.random.RandomState(0)
df = pd.DataFrame({'key': ['A', 'B', 'C', 'A', 'B', 'C'],
                   'data1': range(6),
                   'data2': rng.randint(0, 10, 6)},
                   columns = ['key', 'data1', 'data2'])
df
```

##### (1) 累计

```python
df.groupby('key').mean()
```

```python
df.groupby('key').aggregate(['min', np.median, max])
```

通过 Python 字典指定不同列需要累计的函数

```python
df.groupby('key').aggregate({'data1': 'min',
                             'data2': 'max'})
```

##### (2) 过滤

过滤操作可以按照分组的属性丢弃若干数据

```python
def filter_func(x):
    return x['data2'].std() > 4

print(df); print('-'*20);
print(df.groupby('key').std());print('-'*20);
print(df.groupby('key').filter(filter_func))
```

##### (3) 转换

```python
print(df);print(df.groupby('key').mean())
```

```python
df.groupby('key').transform(lambda x: x - x.mean())
```

##### (4) apply() 方法

apply() 方法可以在每个组上应用任意方法

```python
# 用 apply() 方法将第一列数据以第二列的和为基数进行标准化

def norm_by_data2(x):
    # x 是一个分组的 DataFrame
    x['data1'] /= x['data2'].sum()
    return x

print(df); print(df.groupby('key').apply(norm_by_data2))
```

#### 4.设置分割的键

##### （1）将列表、数组、Series 或索引作为分组键

```python
L = [0, 1, 0, 1, 2, 0]
print(df); print(df.groupby(L).sum())
```

```python
print(df); print(df.groupby(['key']).sum())
```

##### (2) 用字典或 Series 将索引映射到分组名称

```python
df2 = df.set_index('key')  # 将 'key' 设为索引
mapping = {'A': 'vowel', 'B': 'consonant', 'C': 'consonant'}
print(df); print(df2.groupby(mapping).sum())
```

##### (3) 任意 Python 函数

```python
print(df); print(df2.groupby(str.lower).mean())
```

##### (4) 多个有效键构成的列表

任意有效的键都可以组合起来进行分组，从而返回一个多级索引的分组结果

```python
mapping = {'A': 'vowel', 'B': 'consonant', 'C': 'consonant'}
df2.groupby([str.lower, mapping]).mean()
```

#### 5.分组案例

运用上述知识，获取不同方法和不同年份发现的行星数量

```python
decade = 10 * (planets['year'] // 10)  # number // 10 : 抹掉 number 的个位
decade.head()
```

```python
decade = decade.astype(str) + 's'
decade.unique()
```

```python
decade.name = 'decade'
```

```python
planets.head()
```

```python
planets.groupby(['method', decade])['number'].sum().unstack().fillna(0)
```
